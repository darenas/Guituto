# Plotting DL learning curves

In this notebook we use pyplot to plot some curves in a logarithmic scale.<br>
Instead of plotting dummy curves, we are going to use the Learning Curves (LC) for relevant Deep Learning domains. <br>
These LCs were empirically obtained in a research [study](https://arxiv.org/pdf/1712.00409.pdf) performed by members of the [Baidu](www.baidu.com) research team. <br>
The results of the study suggest that the performance of DL algorithms, in terms of reduction of the [generalization error](https://en.wikipedia.org/wiki/Generalization_error), may improve following a power-law as more training data is fed to the model.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(start=1.0, #np.float_power(2,20), 
                stop=np.float_power(2,40), 
                num=np.power(2,18)) # this creates 2^17 numbers evenly spaced between 2^20 and 2^35

plt.figure(figsize=(10,6))
plt.yscale('log')
plt.xscale('log')

plt.plot(X, 21.5 * np.float_power(X,-0.30) + 0.32, 
    label = r'NMT (LSTM-512) - $f(x)=21.2x^{-0.30} + 0.32$') # learning curve for Neural Machine Translation algorithms (LSTM) 

plt.plot(X, 2.24 * np.float_power(X,-0.31),
    label = r'CV (ResNet) - $f(x)=2.24x^{-0.31}$') # learning curve for computer vision (ResNet - scaling model size)

#TODO: add the remaining curves discussed in the paper

plt.legend()
plt.show()
```
